7. Investiga en internet como usar la libreria `re` y crea una funcion que me devuelva en una lista los meteoritos cuyo nombre empieza con una vocal

```python
import re
```

```python
def nombres_con_vocal(columna):
    
    # Expresión regular para nombres que comienzan con una vocal
    inicio_vocal = r'^[aeiouAEIOU]'
    
    nombres_vocal = []
    
    for name in columna:
        if re.match(inicio_vocal, name):
            nombres_vocal.append(name)
    
    return nombres_vocal
```

```python
nombres_con_vocal(df['name'])
```
